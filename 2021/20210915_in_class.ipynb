```python
import nltk
from nltk.corpus import shakespeare
from xml.etree import ElementTree
import os
from bs4 import BeautifulSoup
```

```python
files = os.listdir("shakespeare")
```

```python
word_count = dict()
```

```python
for file in files :
    play_text = open("shakespeare/" + file).readlines()
    clean_play_text = BeautifulSoup(" ".join(play_text), 'html.parser')
    clean_play_list = clean_play_text.text.split()
    clean_play_list = [w.lower() for w in clean_play_list if w.isalpha()]
    
    word_count[file] = [len([w for w in clean_play_list if w == "the"]),
                        len(clean_play_list)]
```

```python
word_count
```

```python
for play, stats in word_count.items() :
    word_count[play].append(stats[0]/stats[1])
    
```
